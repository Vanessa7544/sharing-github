```python
## Step3-Write a query that returns summary statistics for avg_math_4_score by state.
# Make sure to sort the results alphabetically by state name.
sql = '''
SELECT ROUND (AVG(avg_math_4_score), 2 ) AS avg_math_4_score, 
COUNT(state) AS count_avg_math_4_score_record, 
MAX(avg_math_4_score) AS max_avg_math_4_score,
MIN(avg_math_4_score) AS min_avg_math_4_score, state
FROM naep
GROUP BY state, avg_math_4_score
ORDER BY state ;   
'''
stats = engine.execute(sql)
engine.dispose()
avg_math_4_score = stats.fetchall()
print(avg_math_4_score)
```

```python
## Step4-Write a query that alters the previous query so that it returns only the summary statistics for 
# avg_math_4_score by state with differences in max and min values that are greater than 30.
sql = '''
SELECT 
MAX(avg_math_4_score) AS max_avg_math_4_score,
MIN(avg_math_4_score) AS min_avg_math_4_score, state
FROM naep
GROUP BY state
ORDER BY (MAX(avg_math_4_score)-MIN(avg_math_4_score)) > 30;
'''
stats = engine.execute(sql)
engine.dispose()
avg_math_4_score = stats.fetchall()
print(avg_math_4_score)
```

```python
## Step5-Write a query that returns a field called bottom_10_states that lists the states in the bottom 10 for 
# avg_math_4_score in the year 2000.
sql = '''
SELECT state AS bottom_10_state, avg_math_4_score
 FROM naep
 where year = 2000
 ORDER BY avg_math_4_score ASC
 LIMIT 10;
'''
stats = engine.execute(sql)
engine.dispose()
bottom_10_states = stats.fetchall()
print(bottom_10_states)
```

```python
# Step6-Write a query that calculates the average avg_math_4_score rounded to the nearest 2 decimal places 
# over all states in the year 2000.
sql = '''
SELECT 
ROUND (AVG(avg_math_4_score), 2 ) avg_math_4_score 
FROM naep
where year = 2000;
'''
stats = engine.execute(sql)
engine.dispose()
rows = stats.fetchall()
print(rows)
```

```python
## Step7-Write a query that returns a field called below_average_states_y2000 that lists all states with an avg_math_4_score
# less than the average over all states in the year 2000.
sql = '''
SELECT avg_math_4_score AS below_average_states_y2000, year
FROM naep
where avg_math_4_score < 238.92
AND year = 2000;
'''
stats = engine.execute(sql)
engine.dispose()
below_average_states_y2000 = stats.fetchall()
print(below_average_states_y2000)
```

```python
## Step8-Write a query that returns a field called scores_missing_y2000 that lists any states with missing values 
# in the avg_math_4_score column of the naep data table for the year 2000.
sql = '''
SELECT avg_math_4_score AS scores_missing_y2000, year, state
FROM naep
where avg_math_4_score IS NULL 
AND YEAR = 2000;
'''
stats = engine.execute(sql)
engine.dispose()
scores_missing_y2000 = stats.fetchall()
print(scores_missing_y2000)
```
